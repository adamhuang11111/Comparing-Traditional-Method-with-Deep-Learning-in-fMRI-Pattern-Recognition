```python
import numpy as np
import pandas as pd
import os
import seaborn as sns
```

```python
phenotype = pd.read_csv("adhd200_preprocessed_phenotypics.tsv", sep = '\t')
```

```python
phenotype
```

```python
beijing_subjects = os.listdir('anat')
```

```python
def get_list_subject(subjects):
    for i in range(len(subjects)):
        subjects[i] = [int(s) for s in subjects[i].split("_") if s.isdigit()]
        if(len(subjects[i]) > 0):
            subjects[i] = subjects[i][0]
    subjects = [x for x in subjects if x != []]
    return subjects
```

```python
beijing_subjects = get_list_subject(beijing_subjects)
```

```python
beijing_phenotype = phenotype[phenotype["ScanDir ID"].isin(beijing_subjects)]
beijing_phenotype
```

```python
beijing_phenotype.to_csv("beijing_phenotype.csv")
```
